```python
# 统计不同蛋白酶的数量
protease_count = data.groupby([1, 2]).count()[[0]].reset_index()
protease_count.columns = ['Protease', 'substrate', 'Count']
protease_count.head(5)
```

```python
# 查询蛋白酶和底物对是否能保证肽链一致
map_ps_peptide = {}
for row_i, row in data.iterrows():
    protease = row[1]
    substrate = row[2]
    peptide_chain = tuple(row[4:12].tolist())

    # 构建字典
    ps_pair = (protease, substrate)
    if ps_pair not in map_ps_peptide:
        map_ps_peptide[ps_pair] = {peptide_chain}
    else:
        map_ps_peptide[ps_pair] |= {peptide_chain}
```

```python
for x in map_ps_peptide:
    if len(map_ps_peptide[x]) != 1:
        print(x)
```
